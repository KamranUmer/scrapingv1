```python
from bs4 import BeautifulSoup
import requests

# URL of the page
url = 'https://healthwire.pk/doctors/lahore'
# Make a request to fetch the content
page = requests.get(url)
# Parse the page content
soup = BeautifulSoup(page.text, 'html.parser')

# Find the div containing the cities
doctor_listing = soup.find_all("div", class_="doctor_listing_page_wraper_outer")

# Initialize an empty list to store city option elements
city_options = []

# Loop through the found divs (there should be only one in this case)
for listing in doctor_listing:
    # Find the select element for cities
    city_select = listing.find("select", id="city_filter")
    if city_select:
        # Find all option elements within the select
        options = city_select.find_all("option")
        # Add the option elements to the list, skipping the first one (which is 'All Cities')
        city_options.extend(options[1:])

# Create a new BeautifulSoup object with only the city options
city_soup = BeautifulSoup("<select></select>", 'html.parser')
select_tag = city_soup.select
for option in city_options:
    select_tag.append(option)

# Print the prettified version of the city options
print(city_soup.prettify())

```
